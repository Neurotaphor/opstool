# Quick Model and Eigen Visualization

## Plotly-based Visualization

### Model Geometry

```python
import opstool as opst
import opstool.vis.plotly as opsvis
```

```python
opst.load_ops_examples("ArchBridge2")
# or your model code here
```

```python
opsvis.set_plot_props(point_size=0, line_width=3)
```

```python
fig = opsvis.plot_model(show_outline=True)
# fig.show(renderer="jupyterlab")
# fig.show(renderer="notebook")
# fig.show(renderer="browser")
# fig.show()  # for auto
# fig.write_html("**.html")
```

```python
fig.write_html("images/model_plotly.html",
               full_html=False,
               include_plotlyjs="cdn")
```

### Eigen

```python
fig = opsvis.plot_eigen(mode_tags=5, subplots=True, show_outline=False)
# fig.show(renderer="jupyterlab")
# fig.show(renderer="notebook")
# fig.show(renderer="browser")
# fig.show()  # for auto
# fig.write_html("**.html")
```

```python
fig.write_html("images/eigen_subplots_plotly.html",
               full_html=False,
               include_plotlyjs="cdn")
```

```python
fig = opsvis.plot_eigen(mode_tags=5, subplots=False, show_outline=False)
# fig.show(renderer="jupyterlab")
# fig.show(renderer="notebook")
# fig.show(renderer="browser")
# fig.show()  # for auto
```

```python
fig.write_html("images/eigen_slides_plotly.html",
               full_html=False,
               include_plotlyjs="cdn")
```

## Pyvista-based Visualization

```python
import opstool as opst
import opstool.vis.pyvista as opsvis
```

```python
opst.load_ops_examples("Frame3D")
# or your model code here
```

### Model Geometry

```python
opsvis.set_plot_props(point_size=0, line_width=3, notebook=True)

plotter = opsvis.plot_model()
plotter.show(jupyter_backend="jupyterlab")
# plotter.show()
```

### Eigen 

```python
plotter = opsvis.plot_eigen(
    mode_tags=5,
    subplots=True,
)
plotter.show(jupyter_backend="jupyterlab")
# plotter.show()
```
